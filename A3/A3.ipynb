```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_data(filename):
    sizes = []
    types = []
    times = []
    with open(filename, 'r') as f:
        for line in f:
            size, type, time = map(float, line.split())
            sizes.append(size)
            types.append(int(type))
            times.append(time)
    return np.array(sizes), np.array(types), np.array(times)
```

```python
intro_sizes, intro_types, intro_times = read_data("./data/intro.txt")
quick_sizes, quick_types, quick_times = read_data("./data/quick.txt")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(intro_sizes[intro_types == 0], intro_times[intro_types == 0], label="Intro Sort")
plt.plot(quick_sizes[quick_types == 0], quick_times[quick_types == 0], label="Quick Sort")

plt.title("Случайные массивы")
plt.xlabel("Размер массива")
plt.ylabel("Время (микросекунды)")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(intro_sizes[intro_types == 1], intro_times[intro_types == 1], label="Intro Sort")
plt.plot(quick_sizes[quick_types == 1], quick_times[quick_types == 1], label="Quick Sort")

plt.title("Отсортированные массивы")
plt.xlabel("Размер массива")
plt.ylabel("Время (микросекунды)")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(intro_sizes[intro_types == 2], intro_times[intro_types == 2], label="Intro Sort")
plt.plot(quick_sizes[quick_types == 2], quick_times[quick_types == 2], label="Quick Sort")

plt.title("Почти отсортированные массивы")
plt.xlabel("Размер массива")
plt.ylabel("Время (микросекунды)")
plt.legend()

plt.tight_layout()
plt.show()
```

## Вывод

### Случайные массивы

Intro Sort и Quick Sort демонстрируют схожую производительность. Небольшое преимущество может наблюдаться у Intro Sort, особенно на больших размерах массивов. Это связано с тем, что Intro Sort избегает худшего случая $O(n^2)$ для Quick Sort, переключаясь на heapsort при большой глубине рекурсии.

### Отсортированные массивы

Intro Sort значительно превосходит Quick Sort по производительности. Quick Sort в худшем случае (на уже отсортированных данных) деградирует до $O(n^2)$, что хорошо видно на графике. Intro Sort, благодаря своей адаптивности, справляется с отсортированными массивами гораздо эффективнее.

### Почти отсортированные массивы

Intro Sort также показывает преимущество над Quick Sort, хотя разница менее выражена, чем в случае отсортированных массивов. И снова, это объясняется способностью Intro Sort адаптироваться к характеристикам входных данных и предотвращать деградацию производительности.

#### Вывод вывода
В целом, Intro Sort представляет собой более робастный и предсказуемый алгоритм сортировки по сравнению с Quick Sort, избегая худших случаев и обеспечивая гарантированную временную сложность $O(n log_n)$. Небольшое преимущество Quick Sort на случайных массивах может быть связано с меньшими накладными расходами по сравнению с более сложной логикой Intro Sort, но это преимущество нивелируется на отсортированных и почти отсортированных данных.
