```python
data = open("A1.txt").read().split("\n")
data = [list(map(float, i.split(" "))) for i in data]
```

```python
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
from math import pi, asin

res = 0.25 * pi + 1.25 * asin(0.8) - 1
```

```python
from matplotlib import pyplot as plt

plt.plot(x, y)
plt.hlines(res, xmin=0, xmax=100000, colors='r', linestyles='--')
```

```python
import numpy as np

y_1 = [i - res for i in y]
y_2 = [abs(i) / res for i in y_1]

y = np.mean(y, axis=0)
y_std = np.std(y, axis=0)
```

```python
plt.plot(x, y_1)
plt.fill_between(x, y_1 - y_std, y_1 + y_std, alpha=0.2)
```

```python
plt.plot(x, y_2)
```
